# pile range 
224 - 226

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
N_group = 81
T = 10
T2 = 1
Left_edge = 224
Right_edge = 226
```

```python
def data_load(path,N_group):
    f = open(path)
    lines = f.read()
    f.close()
    N_group = N_group
    lines1 = lines.split("\n",(1000*T+1)*N_group-1)
    data_list =[]
    for element in lines1:
        temp = element.split(' ',1)
        temp0 = temp[1]
        temp1 = temp0.strip()
        data_list.append(float(temp1))
    data = np.array(data_list)
    return data
```

# with_a_plie

### u_disp_result_with_pile_210_230.txt

```python
# 'u_disp_210_230_total41_with_a_plie.txt'
u_disp_210_230_total41_with_a_plie1 = data_load('u_disp_result_with_pile_210_230.txt',N_group)
u_disp_210_230_total41_with_a_plie2 = u_disp_210_230_total41_with_a_plie1.reshape(N_group,1000*T+1)
# 1200*41
u_disp_210_230_total41_with_a_plie3 = np.transpose(u_disp_210_230_total41_with_a_plie2)
# delete time 0
u_disp_210_230_total41_with_a_plie = u_disp_210_230_total41_with_a_plie3[1:int(1000*T2+1),:]
# test
plt.plot(u_disp_210_230_total41_with_a_plie[:,1])
```

# Without a pile

### u_disp_result_without_pile_210_230_200N.txt

```python
# 'u_disp_210_230_total41_without_a_plie.txt'
# N_group = 81
u_disp_210_230_total41_without_a_plie1 = data_load('u_disp_result_without_pile_210_230.txt',N_group)
u_disp_210_230_total41_without_a_plie2 = u_disp_210_230_total41_without_a_plie1.reshape(N_group,1000*T+1)
# 1200*41
u_disp_210_230_total41_without_a_plie3 = np.transpose(u_disp_210_230_total41_without_a_plie2)
# delete time 0
u_disp_210_230_total41_without_a_plie = u_disp_210_230_total41_without_a_plie3[1:int(1000*T2+1),:]
# test
plt.plot(u_disp_210_230_total41_without_a_plie[:,1])
```

## Maximum Amplitude

```python
# Maximum Amplitude
# u_disp_210_230_total41_without_a_plie
app4 = []# postive max Amplitude
app4_1 = [] # negative max Amplitude
app4_3 = []# maximum abs Amplitude
app4_2 = [] # postive - negative
for i in range(u_disp_210_230_total41_without_a_plie.shape[1]):
    AA =  (u_disp_210_230_total41_without_a_plie[:,i])
    app4.append(AA.max()) # postive max Amplitude
    app4_1.append(AA.min()) # negative max Amplitude
    abs_AA = abs(AA)
    app4_3.append(abs_AA.max())# maximum abs Amplitude
app4_2 = np.array(app4) - np.array(app4_1) # postive - negative

# u_disp_210_230_total41_with_a_plie
app5 = [] # postive max Amplitude
app5_1 = []# negative max Amplitude
app5_3 = []# maximum abs Amplitude
app5_2 = [] # postive - negative

for i in range(u_disp_210_230_total41_with_a_plie.shape[1]):
    AA = (u_disp_210_230_total41_with_a_plie[:,i])
    app5.append(AA.max())# postive
    app5_1.append(AA.min())# negative
    abs_AA = abs(AA)
    app5_3.append(abs_AA.max())# maximum abs Amplitude
app5_2 = np.array(app5) -np.array(app5_1)# postive - negative
```

## compensation factor

```python
com = [] # compensation
com2 = []
for ele in app4_3:
    com.append((1/ele))

for ele in app5_3:
    com2.append(1/ele)
```
